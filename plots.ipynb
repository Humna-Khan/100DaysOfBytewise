Create a simple line plot using Matplotlib to display the trend of a list of values over time.


```python
import matplotlib.pyplot as plt
time = [2, 6, 8, 10]
values = [33, 41, 66, 85]

plt.plot(time, values, linestyle='-', marker='o')

plt.title('Line for the trends of values over time')
plt.xlabel('Time')
plt.ylabel('Values')
plt.show()
```

Plot a bar chart using Matplotlib to show the frequency of different categories in a dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
sexFreq= df['Sex'].value_counts()


sexFreq.plot(kind='bar', color=['blue', 'yellow', 'red'])

plt.title('Frequency of Gender')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.show()
```

 Create a scatter plot using Matplotlib to visualize the relationship between two variables in a dataset.



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
print(df.head)
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Pclass'], alpha=0.5)

plt.title('Relationship Between Age and Fare of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Pclass')

plt.show()
```

 Load a dataset using Seaborn's built-in dataset functions and create a pairplot to visualize the relationships between all pairs of features
 

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
print(df.head)

sns.pairplot(df, hue='sepal_length', markers=["o", "s", "D"])

plt.suptitle('Pairplot', y=1.06)
plt.show()
```

 Create a box plot using Seaborn to show the distribution of values for different categories in a dataset.


```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('titanic.csv')
print(df.head)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=df)

plt.title('Distribution of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.show()
```

Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('titanic.csv')

# Select only the numeric columns for the correlation matrix
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_columns.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add titles and labels
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```

 Use Matplotlib to create a subplot grid that displays multiple charts in a single figure.




```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('iris')
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    ax = axes[i // 2, i % 2]
    ax.hist(df[feature], bins=10, edgecolor='white', alpha=0.7)
    ax.set_title(f'Distribution of {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Customize the appearance of a Seaborn plot by changing the color palette, adding titles, and modifying axis labels.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

customPalette = ['#FF6349', '#6A4ACD', '#32CD32']

plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, hue='species', palette=customPalette, s=100, alpha=0.8)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

plt.legend(title='Species', loc='upper left')
plt.show()
```

Create a violin plot using Seaborn to visualize the distribution of a dataset across different categories.


```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

plt.figure(figsize=(10, 4))
sns.violinplot(x='sepal_width', y='sepal_length', data=df, palette='Set2')

plt.title('Distribution of Sepal Length Across Iris Species')
plt.xlabel('Sepal Widhth')
plt.ylabel('Sepal Length')
plt.show()
```

Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.


```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

fig, ax = plt.subplots(figsize=(10, 6))

sns.histplot(df['sepal_width'], bins=15, kde=True, color='yellow', ax=ax)

ax.set_title('Histogram')
ax.set_xlabel('Sepal Width')
ax.set_ylabel('Frequency / Density')
plt.show()
```
