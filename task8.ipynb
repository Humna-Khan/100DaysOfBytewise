Calculate the mean, median, and mode of the sepal lengths in the Iris dataset.



```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from scipy import stats

irisData = load_iris()
df = pd.DataFrame(data=irisData.data, columns=irisData.feature_names)
print(df.head)
sepalLength = df['sepal length (cm)']
meanSepalLength = np.mean(sepalLength)
print("Mean of Sepal Length= ", meanSepalLength)
medianSepalLen = np.median(sepalLength)
print("Median of Sepal Length= ", meanSepalLength)
modeSepalLen = stats.mode(sepalLength)
print("Mode of Sepal Length= ", modeSepalLen)


```

Calculate the variance and standard deviation of the petal widths in the Iris dataset.


```python
petalWidth = df['petal width (cm)']
varOfPetalWidth = np.var(petalWidth)
print("Variance of petal width= ", petalWidth)
stdOfPetalWidth = np.std(petalWidth)
print("Standard Deviation of Petal Width= ", stdOfPetalWidth)
```

Define a random variable for the sepal length and calculate the probability distribution of sepal lengths.

Plot the probability distribution of sepal lengths using a histogram.


```python
import scipy.stats as stats
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

irisData = load_iris()
df = pd.DataFrame(data=irisData.data, columns=irisData.feature_names)
print(df.head())

sepalLength = df['sepal length (cm)']

plt.figure(figsize=(10, 8))
plt.hist(sepalLength, bins=8, density=True, color='orange', edgecolor='white')
plt.title("Probability Distribution of Sepal Length")
plt.xlabel('Sepal Length in cm')
plt.ylabel('Probability')
plt.show()
```

Calculate the cumulative distribution function (CDF) for the petal lengths and plot it.



```python
import scipy.stats as stats
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

irisData = load_iris()
df = pd.DataFrame(data=irisData.data, columns=irisData.feature_names)
print(df.head())

petalLength = df['petal length (cm)']

parameters = stats.norm.fit(petalLength)
normalDist=stats.norm(*parameters)
x = np.sort(petalLength)
cdfFordataPoints = normalDist.cdf(x)

plt.figure(figsize=(10,8))
plt.plot(x, cdfFordataPoints,linestyle='-', color='green')
plt.title("Cumulative Distribution Function for Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Cummulative Distribution")
plt.grid(True)
plt.show()
```

Calculate and plot the probability density function (PDF) for sepal width.


```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from scipy.stats import gaussian_kde
import numpy as np

irisData = load_iris()
df = pd.DataFrame(data=irisData.data, columns=irisData.feature_names)
sepalWidth = df['sepal width (cm)']

kde = gaussian_kde(sepalWidth)
xValues = np.linspace(min(sepalWidth), max(sepalWidth), 1000)
pdfValues = kde.evaluate(xValues)

plt.figure(figsize=(8, 6))
plt.plot(xValues, pdfValues, color='orange')

plt.title('Probability Density Function (PDF) of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')

plt.show()
```
